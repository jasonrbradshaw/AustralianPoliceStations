```python
import urllib.request
path = 'https://www.police.wa.gov.au/Contact-Us/Police' 
with urllib.request.urlopen(path) as response:
    webSite = response.read().decode('utf-8')
```

```python
# Print HTML from site to locate strings
print(webSite)
```

```python
import re
```

```python
# Remove tab character from links
webSite = re.sub(r'\t',"", webSite)
```

```python
# Put station names into a variable
names = re.findall(r'<a href="https://www.police.wa.gov.au/Contact%20Us/LPT.*?">(.*?)<',webSite)
```

```python
# Put links to station sites in a variable for later
link = re.findall(r'<a href="(https://www.police.wa.gov.au/Contact%20Us/LPT.*?)">',webSite)
```

```python
print(link)
```

```python
# Replace blank space in links so they load properly
links = []

for l in link:
    data = re.sub(r' ',"%20", l)
    
    links.append(data)
```

```python
print(links)
```

```python
# Function to extract station locations from external sites
def StationLocation(url):
    location = []
    
    loc = re.findall(r'<span>(.*?)</span><br>', url)
    
    location = loc
    
    return location;
```

```python
# Function to extract Twitter handles from external sites
def TwitterHandle(url):
    twitter = []
    
    twit = re.findall(r'Tweets by (.*?)<', url)
    
    twitter = twit
    
    return twitter;
```

```python
# For-loop to extract location data from stored links
location=[]

for l in links:
    url = l
    with urllib.request.urlopen(url) as response:
        url = response.read().decode('utf-8')
    
    data = StationLocation(url)
    
    location.append(data)
```

```python
# For-loop to extract Twitter handle data from stored links
twitter=[]

for l in links:
    url = l
    with urllib.request.urlopen(url) as response:
        url = response.read().decode('utf-8')
    
    data = TwitterHandle(url)
    
    twitter.append(data)
```
